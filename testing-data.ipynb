What country saw the biggest 5 year change in net migration?

Which country saw the biggest swing in net migration from 1962 to 2017?

What economic indicators drive immigration or emigration?

Which continents see the most immigration or emigration?

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

net_migration = pd.read_csv('net-migration-data.csv')
net_migration.head()
```

```python
cleaner_migration = net_migration[['Country Name', 'Country Code', '1962', '1967', '1972', '1977', '1982', '1987', '1992', '1997', '2002', '2007', '2012', '2017']]
```

```python
#cleaner_migration.info()
cleaner_migration = cleaner_migration.dropna()
cleaner_migration.head()
```

```python
#Isolating economic and aggregate columns
econ_factors = ['Fragile and conflict affected situations','High income','Heavily indebted poor countries (HIPC)',
                'Least developed countries: UN classification','Low income','Lower middle income','Low & middle income',
                'Middle income','Upper middle income']
continents = ['North America', 'Central Europe and the Baltics', 'Australia', 'East Asia & Pacific',
              'Africa Eastern and Southern','Africa Western and Central','Europe & Central Asia','European Union',
             'Latin America & Caribbean','Middle East & North Africa','Sub-Saharan Africa',]

continent_migration = cleaner_migration[cleaner_migration['Country Name'].isin(continents)]
econ_migration = cleaner_migration[cleaner_migration['Country Name'].isin(econ_factors)]
```

```python
#This shows net change and removes most year cols
net_migration = cleaner_migration.copy()
net_migration['Net Change'] = net_migration.apply(lambda row: row['2017'] - row['1962'], axis = 1)
net_migration = net_migration[['Country Name', '1962', '2017', 'Net Change']]
```

```python
#Greatest negative net change
net_migration.sort_values('Net Change')[12:13]
```

```python
#Greatest positive net change
net_migration.sort_values('Net Change')[233:234]
```

```python
net_migration_continent = net_migration[net_migration['Country Name'].isin(continents)]
net_migration_continent = net_migration_continent[['Country Name', 'Net Change']]
net_migration_continent = net_migration_continent.sort_values('Net Change')
#ax = net_migration_continent.plot(title = 'Change in Net Migration by Continent',kind='bar')
#ax.set_xlabel('Continent')
#ax.set_ylabel('Change in Net Migration from 1962 to 2017')
#plt.show()
plt.figure(figsize=(10,10))
plt.bar(net_migration_continent['Country Name'],net_migration_continent['Net Change']) 
plt.xticks(rotation = 90)
plt.title("Change in Net Migration by Continent from 1962 to 2017")
plt.xlabel("Continent")
plt.ylabel("Net Change in Migration")
plt.show()
```

```python
net_migration_econ = net_migration[net_migration['Country Name'].isin(econ_factors)]
net_migration_econ = net_migration_econ[['Country Name', 'Net Change']]
net_migration_econ = net_migration_econ.sort_values('Net Change')
```

```python
plt.figure(figsize=(10,10))
plt.bar(net_migration_econ['Country Name'],net_migration_econ['Net Change']) 
plt.title("Change in Net Migration by Economic Factors from 1962 to 2017")
plt.xlabel("Economic Factor")
plt.xticks(rotation = 90)
plt.ylabel("Net Change in Migration")
plt.show()
```
